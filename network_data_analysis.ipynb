From here we will analyse the karate club graph network which is given y the object karate

```python
karate=nx.read_gml('karate.gml',label='id')
```

```python
possible_degrees=list(set(dict(nx.degree(karate)).values()))
```

```python
def plot_deg_dist(A):
    all_degrees=list(dict(nx.degree(A)).values())
    possible_degrees=list(set(dict(nx.degree(A)).values()))
    count=[]
    for i in possible_degrees:
        x=all_degrees.count(i)
        count.append(x)
    plt.plot(possible_degrees,count,'yo-')
    plt.xlabel('degrees')
    plt.ylabel('degree distribution')
    plt.show()
```

```python
plot_deg_dist(karate)
```

IN general normal graphs follow power law distribution.the loglog plot will be linear in such case.

```python
def plot_deg_power_law(A):
    all_degrees=list(dict(nx.degree(A)).values())
    possible_degrees=list(set(dict(nx.degree(A)).values()))
    count=[]
    for i in possible_degrees:
        x=all_degrees.count(i)
        count.append(x)
    plt.loglog(possible_degrees,count,'yo-')
    plt.xlabel('degrees')
    plt.ylabel('degree distribution')
    plt.show()
```

```python
plot_deg_power_law(karate)
```
